```python
#문자열 메소드
import pandas as pd
import numpy as np

ser=pd.Series(['Apple_사과','Banana_바나나','Cherry_체리',np.nan],index=["First "," Second"," Third", "Fourth"])
ser2=pd.Series(['Apple_사과','Banana_바나나','Cherry_체리',np.nan],index=["First "," Second"," Third", "Fourth"],dtype='string')
print(ser)
print(ser2)
```

```python
#소문자로 변환
print(ser.str.lower())
print("\n")
#대문자로 변환
print(ser.str.upper())
print("\n")
#문자열 길이 계산
print(ser.str.len())
```
