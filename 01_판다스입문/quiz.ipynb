```python
data = np.array([1, 2, 3, 4, 5])
# 14. data 배열에서 2보다 큰 원소들만 추출하세요.
cond=data>2
data[cond]
```

```python
# 15. data 배열에서 2보다 크면서 동시에 5보다 작은 원소들만 추출하세요. (비트 연산자 & 활용)
cond=(data>2) & (data<5)
print(data[cond])
```

```python
stats_arr = np.array([1, 2, 3, 4])
# 16. 합계, 평균, 분산, 표준편차, 중앙값을 구하세요
print(stats_arr.sum()) #합계
print(np.sum(stats_arr)) #합계
print(np.mean(stats_arr)) #평균
print(np.var(stats_arr)) #분산
print(np.std(stats_arr)) #표준편차
print(np.median(stats_arr)) #중앙값
```
