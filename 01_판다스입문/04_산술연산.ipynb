```python
#시리즈 시리즈 연산
student1=pd.Series({
    "국어":100,
    "영어":80,
    "수학":90
})
student2=pd.Series({
    "수학":80,
    "국어":90,
    "영어":80,  
})
print(student1)
print(student2)
```

```python
#순서가 달라도 같은 국어, 수학, 영어 점수를 더함. (/*-+ 모두 가능)
print(student1+student2)
print(type(student1+student2))
```

```python
#Nan값이 있는 시리즈 연산 (Nan=값이 없음)
import pandas as pd
import numpy as np
student1=pd.Series({'국어':np.nan,'영어':80,'수학':90})
student2=pd.Series({'수학':80,'국어':90})
print(student1)
print(student2)
```

```python
add=student1+student2
sub=student1-student2
mul=student1*student2
div=student1/student2
print(add, sub, mul, div)
```

```python
result=pd.DataFrame([add,sub,mul,div],index=['덧셈','뺄셈','곱셈','나눗셈'])
result
```

```python
#연산 메소드
#두 사람의 과목별 점수로 사칙연산
sr_add=student1.add(student2,fill_value=0)
sr_sub=student1.sub(student2,fill_value=0)
sr_mul=student1.mul(student2,fill_value=0)
sr_div=student1.div(student2,fill_value=0)
result=pd.DataFrame([sr_add,sr_sub,sr_mul,sr_div],index=['덧셈','뺄셈','곱셈','나눗셈'])
result
```

```python
#데이터프레임끼리 더하기
import pandas as pd
import seaborn as sns

titanic=sns.load_dataset('titanic')
df=titanic.loc[:,['age','fare']]
print(df.tail())
print(type(df))
```

```python
add=df+10
print(add.tail)
```

```python
sub=add-df
print(sub.tail())
```

```python
#
sample1=add.tail()
sample2=sub.tail().fillna(0.0)
print(sample1)
print(sample2)
```

```python
df_add=sample1.add(sample2,fill_value=0)
df_sub=sample1.sub(sample2,fill_value=0)
df_mul=sample1.mul(sample2,fill_value=0)
df_div=sample1.div(sample2,fill_value=0)
print(df_add)
print(df_sub)
print(df_mul)
print(df_div)
```
