```python
#불린 인덱싱
import pandas as pd
import seaborn as sns
titanic=sns.load_dataset('titanic')
df=titanic.loc[0:9,['age','fare']]

df
```

```python
#10살 이하인지 판별
df['age']<10
```

```python
#10살 이하인 사람만 필터링
df[df['age']<10]
```

```python
#20세 미만이 아닌 사람만 필터링. "~"는 부정의 의미.
df[~(df['age']<20)]
```

```python
#10대만 출력
mask1=(titanic.age>=10)&(titanic.age<20)
df_teenage=titanic[mask1]
df_teenage.head() #head()는 첫 5행만 출력하는 것.
```

```python
#10대 여성만 출력
mask2=(titanic.age<10)&(titanic.sex=='female')
df_female_under10=titanic.loc[mask2]
df_female_under10.head()
```

```python
#10세미만 60세이상의 사람들을 'age','sex','alone' 만 출력
mask3=(titanic.age<10)|(titanic.age>=60)
df_under10_morethan60=titanic.loc[mask3,['age','sex','alone']]
df_under10_morethan60.head()
```

```python
#query()메소드
import pandas as pd
import seaborn as sns
titanic=sns.load_dataset('titanic')
df=titanic.loc[0:9,['age','fare']]

df
```

```python
#20세 미만의 조건을 충족하는 행을 필터링
df.query('age<20')
```

```python
#20세 이상이고 요금이 15 미만인 조건을 모두 충족하는 행을 필터링
df.query('age>=20 and fare<15')
```

```python
#나이가 20세 이상 또는 요금이 15 미만인 조건을 하나라도 충족하는 행을 필터링
df.query('age>=20 or fare<15')
```

```python
#isin() 메소드
import pandas as pd
import seaborn as sns
titanic=sns.load_dataset('titanic')
#출력할 최대 열의 개수 10개
pd.set_option('display.max_columns',10)

#탑승한 도시가 사우스햄튼, 퀸스타운인 사람만 출력
mask1=titanic['embark_town']=="Southampton"
mask2=titanic['embark_town']=="Queenstown"
df_boolean=titanic[mask1|mask2]
df_boolean.head()
```

```python
#isin을 활용한 방법
isin_filter=titanic['embark_town'].isin(["Southampton","Queenstown"])
df_isin=titanic[isin_filter]
df_isin.head()
```
