```python
#원소 선택
import pandas as pd

exam_data={
    '이름':['서준','우현','인아'],
    '수학':[90,80,70],
    '영어':[98,89,95],
    '음악':[85,95,100],
    '체육':[100,90,90]
}
df=df.set_index('이름')
df
```

```python
#원소 선택 : 행과 열 선택하기
a=df.loc["서준","음악"]
b=df.iloc[0,2]
print(a,b)
```

```python
#두개 이상의 값 가져오기.
#아래 c d e f 는 모두 같은 값을 불러옴.
c=df.loc['서준',['음악','체육']]
print(c)
d=df.iloc[0,[2,3]]
print(d)
e=df.loc['서준','음악':'체육']
print(e)
f=df.iloc[0,2:]
print(f)
```

```python
#떨어져있는 요소 가져오기
g=df.loc[['서준','우현'],['음악','체육']]
print(g)
h=df.iloc[[0,1],[2,3]]
print(h)
i=df.loc['서준':'우현','음악':'체육']
print(i)
j=df.iloc[0:2,2:]
print(j)
```
